```python
#to open a file
with open("notes.txt","r") as file:
    content=file.read()
print(content)
```

```python
#to write inside a file
with open("notes.txt", "w") as file:
    file.write("Hello, Samrajya\n")
    file.write("Welcome to AI engineering")

#write overwrites everuthing in an existing file
```

```python
#to append
with open("notes.txt", "a") as file:
    file.write("\nToday's Lesson: File Handling")
```

```python
#to read line by line

with open("notes.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
#exercise1

def line_count(file_path):
    with open(file_path, "r")as file:
        return sum(1 for line in file)

total_lines=line_count("notes.txt")
print("Total lines: ",total_lines)

#we can also use len(file.readlines) but it forces python to load every single line in computer's RAM simultaneously. if the size
#is in GB it may cause the computer to crash
```

```python
#exercise2
def words_count(file_path):
    with open(file_path, "r")as file:
        return sum(len(line.split())for lines in file) 

total_words=words_count("notes.txt")
print(total_words)
```

```python
with open("notes.txt","r")as file:
    for lines in file:
        line=len(lines.strip())
        print("Longest line:", max(str(line)))
```
