```python
#to create a random password

size=int(input("Enter the size of the password u want to generate: "))

chars="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"

password="".join(random.sample(chars,size))
print(password)
```
