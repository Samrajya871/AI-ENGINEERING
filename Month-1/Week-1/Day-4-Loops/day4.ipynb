#LOOPS
for and while loop

```python
for i in range (5):
    print (i+1)
```

```python
for i in range (1,5):
    print (i)
```

```python
for i in range(1,21,2):
    print(i)
```

```python
# while loops run till the condition is true

#example

num=1
while num<=5:
    print(num)
    num+=1
```

```python
num=5
while num>=1:
    print(num)
    num-=1
print("blast off!")
```

```python
#sometimes we want to stop a loop early
#so we use break

#example

while True:
    name=input("Enter Name")

    if name=="exit":
        break
    print("Hello", name)
```

```python
num=5
for i in range(num):
    if i ==4:
        break
    print(i)
```

```python
#now if we want to skip an iteration we use continue
#we use continue

num=5
for i in range(num):
    if i==3:
        continue
    print(i)
```

```python
total = 0
for i in range(1,6):
    total+=i
print(total)
```
