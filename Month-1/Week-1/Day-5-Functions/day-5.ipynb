```python
#IMAGINE WE WANT TO WRITE SOMETHING THRICE
print("Welcome to AI")
#to print this we will have to write it three times
```

```python
#but if we use function

def welcome():
    print("Welcome to AI")

welcome()
welcome()
welcome()
```

```python
#Parameters
#functions become powerful when they accept input

def greet(name):
    print("Hello", name)

greet("sam")
greet("ram")
```

```python
def square(num):
    print(num*num)

square(5)
square(1000)
```

```python
#return values
#we use return value so another part of the program can use it
 
def add(a,b):
    return a+b

result=add(10,20)

print(result)
```

```python
def cube(n):
    return n*n*n

result=cube(5)

print(result)
```
